```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')

iris.head()
```

```python
print("Shape of Dataset:")
print(iris.shape)
```

```python
print("Column Names:")
print(iris.columns)
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species'
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
iris.hist(figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=iris)

plt.title("Box Plot of Iris Features")
plt.show()
```

```python
# Conclusion
'''
- The Iris dataset contains 150 flower samples.
- There are three species: Setosa, Versicolor, and Virginica.
- Petal length and petal width clearly separate the species.
- Box plots help identify spread and possible outliers.
- Histograms show the distribution of feature values.
'''
```
